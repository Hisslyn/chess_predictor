# Chess Strategy Recommender — Test Scenarios

Tests 5 game situations against the trained XGBoost model.

```python
import sys, os
sys.path.insert(0, os.path.abspath('..'))   # repo root
os.chdir(os.path.abspath('..'))             # make model/data paths resolve
from src.recommend import recommend
```

## Scenario 1 — Elite player, round 4/7, leading field, vs stronger with Black

*2500 player, round 4 of 7, score 2.5/3, playing Black vs 2580, leading the field.*

```python
result = recommend(
    player_rating=2500,
    opponent_rating=2580,
    round_num=4,
    n_rounds=7,
    current_score=2.5,
    gap_to_leader=0.0,
    playing_white=0,
    tournament_avg_rating=2350,
    tournament_rating_std=150,
    field_size=40,
    title_encoded=6,
)
# pretty narrative
rec  = result["recommended"].upper()
conf = result["confidence"]
rp_s = result["solid"]["predicted_rp"]
rp_a = result["aggressive"]["predicted_rp"]
rp_p = result["passive"]["predicted_rp"]
exp  = result["explanation"]

print(f"You're 2500, round 4 of 7, 2.5/3 pts, "
      f"vs 2580 with black (leading).")
print(f"Recommendation: {rec}. "
      f"Predicted Rp: solid={rp_s:.0f}, aggressive={rp_a:.0f}, passive={rp_p:.0f}.")
print(f"Confidence: {conf}.")
print(f"Key factors: score_delta ={result['aggressive']['delta_vs_best']:+.1f} vs best, "
      f"gap_to_leader +0.0, rating_diff -80.")
print(f"{exp}")
print()
print("Full result dict:", result)
```
